```python


import datetime as dt
import xarray as xr
import fsspec
import s3fs
import os.path
import matplotlib.pyplot as plt
```

```python
def get_geo_data(sat='goes-west', lyr=2020, idyjl=1, product='ABI-L2-SSTF'):
    # arguments
    # sat   goes-east,goes-west
    # lyr   year
    # idyjl day of year
    # Product. For a full list see: https://github.com/NOAA-Big-Data-Program/nodd-data-docs/tree/main/GOES

    d = dt.datetime(lyr,1,1) + dt.timedelta(days=idyjl)
    fs = s3fs.S3FileSystem(anon=True) #connect to s3 bucket!

    #create strings for the year and julian day
    imon,idym=d.month,d.day
    syr,sjdy,smon,sdym = str(lyr).zfill(4),str(idyjl).zfill(3),str(imon).zfill(2),str(idym).zfill(2)

    #use glob to list all the files in the directory
    if sat=='goes-east':
        file_location,var = fs.glob(f's3://noaa-goes16/{product}/{syr}/{sjdy}/*/*.nc'),'SST'
    if sat=='goes-west':
        file_location,var = fs.glob(f's3://noaa-goes17/{product}/{syr}/{sjdy}/*/*.nc'),'SST'

    #make a list of links to the file keys
    if len(file_location)<1:
        return file_ob
    file_ob = [fs.open(file) for file in file_location]        #open connection to files

    #open all the day's data
    ds = xr.open_mfdataset(file_ob,combine='nested',concat_dim='time') #note file is super messed up formatting

    #clean up coordinates which are a MESS in GOES
    #rename one of the coordinates that doesn't match a dim & should

    ds = ds.rename({'t':'time'})
    ds = ds.reset_coords()

    return ds
```

```python

ds = get_geo_data('goes-east', 2020, 210)
```

```python
subset = ds.sel(x=slice(-0.01, 0.06), y=slice(0.12, 0.09))

# Masking out bad data (clouds/errors) using the Quality Flags
# DQF == 0 is the 'Good' data
masked = subset.SST.where(subset.DQF == 0)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# --- STEP 1: LOAD THE FULL DISK ---
# Removing slices to see the whole hemisphere
full_masked = ds.SST.where(ds.DQF == 0)
full_mean = full_masked.mean('time', skipna=True)

# 1. THE GLOBAL THERMAL GRADIENT (North to South)
plt.figure(figsize=(12, 6))
full_mean.plot(cmap='magma', vmin=270, vmax=305)
plt.title('Graph 1: Western Hemisphere Sea Surface Temperature (Full Disk)')
plt.show()

# 2. GLOBAL THERMAL VOLATILITY (The Energy Map)
# Where is the water changing the most in 24 hours?
plt.figure(figsize=(12, 6))
thermal_range = full_masked.max('time') - full_masked.min('time')
thermal_range.plot(cmap='hot', vmin=0, vmax=4)
plt.title('Graph 2: Hemisphere Diurnal Range (Where is the Sun hitting hardest?)')
plt.show()

# 3. LATITUDINAL TEMPERATURE PROFILE
# This collapses the map into a single line from North to South
plt.figure(figsize=(10, 4))
full_mean.mean(dim='x').plot(color='blue', linewidth=2)
plt.title('Graph 3: Average Temperature vs. Latitude (North to South)')
plt.ylabel('Temp (K)')
plt.grid(True, alpha=0.3)
plt.show()

# 4. DATA QUALITY / CLOUD COVER MAP
# This shows where it was sunny vs cloudy across the hemisphere
plt.figure(figsize=(12, 6))
quality_map = ds.DQF.mean('time')
quality_map.plot(cmap='Blues_r')
plt.title('Graph 4: Cloud Cover & Data Quality (Where can the satellite see?)')
plt.show()

# 5. SPATIAL FRONT DETECTION (The Edges)
# This highlights the Gulf Stream and other major currents
plt.figure(figsize=(12, 6))
dy, dx = np.gradient(full_mean.fillna(full_mean.mean()).values)
grad_mag = np.sqrt(dx**2 + dy**2)
plt.imshow(grad_mag, cmap='viridis', origin='upper', vmax=5)
plt.title('Graph 5: Global Thermal Fronts (Currents & Boundaries)')
plt.colorbar(label='Gradient Magnitude')
plt.show()

# 6. HEMISPHERIC TEMPERATURE DISTRIBUTION
plt.figure(figsize=(8, 5))
data_to_plot = full_masked.values.flatten()
clean_vals = data_to_plot[~np.isnan(data_to_plot)]
sns.histplot(clean_vals, bins=100, color='purple', kde=True)
plt.title('Graph 6: Statistical Distribution of Temperatures (Entire Hemisphere)')
plt.xlabel('Temperature (K)')
plt.show()
```
